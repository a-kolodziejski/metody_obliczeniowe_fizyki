Do rozwiązania zadania wykorzystano metodę Rungego-Kutty czwartego rzędu (Runge-Kutta 4th Order Method), gdzie:

$\ y_{n+1} = y_n + \frac{h}{6} (k_1 + 2k_2 + 2k_3 + k_4)$

$\ k_1 = hf(t_n, y_n)$

$\ k_2 = hf(t_n + \frac{h}{2}, y_n + \frac{k_1}{2})$

$\ k_3 = hf(t_n + \frac{h}{2}, y_n + \frac{k_2}{2})$

$\ k_4 = hf(t_n + h, y_n + k_3)$





```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parametry z zadania
m = 2.5 #kg
k = 75 #N/m
```

```python
#definiujemy funkcję podaną w treści zadania; f = d^2 y/dt^2
def f(t,y,P):
  return P(t)/m - k*y/m
```

```python
#definiujemy funkcję dla siły P(t) przy warunkach podanych w treści zadania
def P(t):
  return 10*t if t<2 else 20
```

```python
#definiujemy funkcję metody Rungego-Kutty
#przez h definiujemy 'krok' - odstęp czasowy, T - całkowity czas
def runge_kutta(h,T):
  t_wart = np.arange(0, T+h, h)
  y_wart = [0]

  for t_n in t_wart[:-1]:
        k1 = h * f(t_n, y_wart[-1], P)
        k2 = h * f(t_n + h/2, y_wart[-1] + k1/2, P)
        k3 = h * f(t_n + h/2, y_wart[-1] + k2/2, P)
        k4 = h * f(t_n + h, y_wart[-1] + k3, P)

        y_nowe = y_wart[-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
        y_wart.append(y_nowe)

  return t_wart, y_wart
```

```python
#wywołanie funkcji i naszkicowanie wykresu dla wybranych h i T

h = 0.01
T = 5

t_wart, y_wart = runge_kutta(h, T)

#wychylenie maksymalne
max_wychylenie = max(y_wart)
print('Wychylenie maksymalne:', max_wychylenie)

#wykres
plt.plot(t_wart, y_wart)
plt.title('Układ z masą na sprężynie')
plt.xlabel('Czas (s)')
plt.ylabel('Wychylenie (m)')
plt.show()
```

**Eksperymenty z wartością h i T**

W metodzie Rungego-Kutty wartości 'kroku' h i ilość iteracji (lub w tym przypadku czas całkowity T) dobiera się metodą prób i błędów w zależności od pożądanej dokładności oraz możliwości obliczeniowych komputera.
